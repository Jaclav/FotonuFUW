**Czyli SPDC I ma mniejszą moc progową względem SPDC II, ale też mniejszą tolerancję kątową**, optymalne kąty:
$$\theta_{I}=31.2^\circ$$
$$\theta_{II}=45.3^\circ$$


## Dla SPDC I
$$e \rightarrow o + o$$
$$\theta_{OPT} = 31.2^\circ (\text{SNLO})$$
$\alpha$ - kąt między jednym produktem SPDC I po wyjściu z kryształu, a normalną do powierzchni kryształu,\
$\beta$ - kąt między drugim produktem SPDC I po wyjściu z kryształu, a normalną do powierzchni kryształu\
Primowane wartości są wewnątrz kryształu\
$$\theta = 29.2^\circ(\text{Crystal specification})$$
![diag1.png](diag1.png)


$$\frac{1}{\lambda_p}=\frac{1}{\lambda_s}+\frac{1}{\lambda_i}=\frac{2}{\lambda_{is}}$$
$$\gamma^\prime+\theta=\theta_{OPT}$$
$$n_{e,eff}(\lambda,\theta)=\left(\frac{\cos^2\theta}{n_o(\lambda)^2}+\frac{\sin^2\theta}{n_e(\lambda)^2}\right)^{-1/2}$$
$$k_{p, inside}=k_p^\prime=n_{e,eff}(\lambda_{p,vac},\gamma^\prime+\theta)\frac{2\pi}{\lambda_{p,vac}}$$
$$\lambda_i^\prime=\lambda_s^\prime \Longrightarrow \alpha^\prime=\beta^\prime$$
$$\vec{k_p^\prime}=\vec{k_s^\prime}+\vec{k_i^\prime} \Longrightarrow k_s^\prime=\frac{k_p^\prime}{2\cos\alpha^\prime} \Longrightarrow \frac{k_p^\prime}{2k_s^\prime} \leq 1$$
$$\sin\gamma=n_{e,eff}(\lambda_p,\gamma^\prime+\theta)\sin\gamma^\prime$$
$$\sin \alpha = n_o(\lambda_{is})\sin(\alpha^\prime + \gamma^\prime)$$

````python
import numpy as np
import matplotlib.pyplot as plt
from engineering_notation import *
from BBO import *
````

````python
lp = 0.370
ls = lp * 2  # idle and signal
THETA = np.radians(29.2)
THETA_OPT = np.radians(32.5)


gp = THETA_OPT - THETA
gamma = np.arcsin(neeff(lp, THETA_OPT) * np.sin(gp))
print("gamma =", np.degrees(gamma))
print("gamma^\prime =", np.degrees(gp))


kp_prime = neeff(lp, THETA_OPT) * 2 * np.pi / lp
kis_prime = no(ls) * 2 * np.pi / ls
assert kp_prime / (2 * kis_prime) <= 1
ap = np.arccos(kp_prime / (2 * kis_prime))
print("alpha^\prime =", np.degrees(ap))
````

Ponieważ $\alpha^\prime < \gamma^\prime$, więc\
![diag2.png](diag2.png)
$$\sin \beta  = n_o(\lambda_{is})\sin(\gamma ^\prime-\beta^\prime)$$
$$\delta=\alpha-\beta$$

````python
bp = ap
alpha = np.arcsin(no(ls) * np.sin(ap + gp))
beta = np.arcsin(no(ls) * np.sin(gp - bp))
print(
    "alpha =",
    np.degrees(alpha),
    "\nbeta =",
    np.degrees(beta),
    "\ndelta =",
    np.degrees(alpha - beta),
    "\ndelta/2 =",
    np.degrees(alpha - beta) / 2,
)
````

Kąty między fotonami wyjściowymi, a wiązką lasera pompującego:

````python
print(np.degrees(alpha - gamma))
print(np.degrees(beta - gamma))
````
